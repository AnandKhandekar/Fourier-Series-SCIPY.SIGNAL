```python
# import depndencies
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal   # sciy is an advanced SCIENTIFIC processing library
from scipy.integrate import simps
```

```python
t = np.linspace(0, 1 , 1000, endpoint=True)
# Plot the readymade SAW tooth shaped signal 
plt.plot( t, signal.sawtooth(4*np.pi*t)) 
plt.xlabel("Time")
plt.ylabel('Amplitude')
plt.title('Readymade saw tooth wave signal')
plt.axhline(y=0,color='g')
plt.show()
```

```python
L = 1
terms = 50 
freq = 2 
width_range = 1 
samples = 1000
```

```python
x = np.linspace( 0, L, samples, endpoint=True)
y = signal.square(2*np.pi*freq*x /L )
plt.plot(x,y)
plt.title('Readymade Saw tooth Wave')
```

#Calculate a0,an and bn

use Lambda function

#Fourier Series Constants
* $\huge a_0 = \frac{2}{L}\;  \int_{c}^{c+2L}f(x)\;dx$

* $\huge a_n = \frac{2}{L}\;  \int_{c}^{c+2L}f(x)\;\;cos(\frac{2n\;\pi\;x}{L})dx$
* $\huge b_n = \frac{2}{L}\;  \int_{c}^{c+2L}f(x)\;\;sin(\frac{2n\;\pi\;x}{L})dx$

```python
a0 = 2.0/ L * simps(y,x) 
an = lambda n : 2.0 / L * simps ( y * np.cos ( 2* n * np.pi * x / L) , x ) 
bn = lambda n : 2.0 / L * simps ( y * np.sin ( 2* n * np.pi * x / L) , x ) 
```

#Fourier Series summation equation
$\huge f(x) = \frac{a_0}{2} + \sum_{n=1}^{\infty}[ a_n\;cos(\frac{n\;\pi\;x}{L})+b_n\;sin(\frac{n\;\pi\;x}{L})]$

substitute the values of a0, an and bn in the Fourier series Formula

```python
s_saw = a0 / 2.0 + sum( [an(k) * np.cos( 2.0 * k * np.pi * x / L )  +  bn(k) * np.sin( 2.0 * k * np.pi * x / L )  for k in range(1, terms +1)]   )  # LIST comprehension
```

```python
plt.figure(figsize=(18,5))

plt.subplot(121)
plt.plot(x,y,label="original sawtooth wave",color='k',linewidth =2)
plt.xlabel("$x$")
plt.ylabel("$y = f(x)$")
plt.legend(loc='best', prop = {'size':10})
plt.title("Readymade/Original Sawtooth wave")

plt.subplot(122)
plt.plot(x,s_saw,label="sawtooth wave using Fourier",color='b',marker='.') 
plt.plot(x,y,label="original sawtooth wave",color='k',linewidth =2)
plt.xlabel("$x$")
plt.ylabel("$y = f(x)$")
plt.legend(loc='best', prop = {'size':10})
plt.title("Sawtooth wave using Fourier Series")
```
